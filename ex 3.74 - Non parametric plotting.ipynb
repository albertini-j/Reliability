```python
#ex 3.74

from sympy import *
init_session()
from sympy.stats import *
import matplotlib.pyplot as plt
import numpy as np


F, a, b, R, mttf, l =symbols('F a b R mttf l')

```

```python
f_time= np.sort(np.array([15.1, 10.7, 8.8, 11.3, 12.6, 14.4,8.7]))
```

```python
#estimate h(ti)
n=len(f_time)
hazard=[]
for i in range(1,len(f_time)):
    hazard.append(round(1/((n-i+0.625)*(f_time[i]-f_time[i-1])),4))
                  
```

```python
plt.plot(f_time[:-1],hazard)
plt.scatter(f_time[:-1],hazard)
plt.xlabel('test time(months)')
plt.ylabel('Hazard rate')
plt.title('Non parametric hazard rate plot')
plt.grid(True)
```

```python
reliability=[]
for i in range(1,len(f_time)+1):
    reliability.append(round((n-i+0.625)/(n+0.25),4))
                  
```

```python
plt.plot(f_time,reliability)
plt.scatter(f_time,reliability)
plt.xlabel('test time (months)')
plt.ylabel('Reliability rate')
plt.title('Non parametric reliability plot')
plt.grid(True)
```

```python
failure_pdf=[]
for i in range(1,len(f_time)):
    failure_pdf.append(round(1/((n+0.25)*(f_time[i]-f_time[i-1])),4))
                  
```

```python
plt.plot(f_time[:-1],failure_pdf)
plt.scatter(f_time[:-1],failure_pdf)
plt.xlabel('test time(months)')
plt.ylabel('Failure pdf')
plt.title('Non parametric failure pdf plot')
plt.grid(True)
```
