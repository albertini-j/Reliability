```python
#Ebeling 9.19
from sympy import *
init_session()
from sympy.stats import *
import matplotlib.pyplot as plt
import numpy as np

F, a, b, R, mttf, l =symbols('F a b R mttf l')

#primary unit:
mttf_p=1000
#standby unit - active:
mttf_s=333
#standby unit - standby:
#mttf_s_=2000 - desconsiderado
#repair rate:
lr=0.01
```

```python
t=symbols('t')
lp=1/mttf_p
ls=1/mttf_s
k1=lp+ls+lr
k2=lp*ls
x1=(-k1+sqrt(k1**2-4*k2))/2
x2=(-k1-sqrt(k1**2-4*k2))/2
R=((k1+x1)*exp(x1*t)-(k1+x2)*exp(x2*t))/(x1-x2)
R
```
